```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
```

```python
x = 6 * np.random.rand(200,1) - 3
y = 0.8 + x**2 + 0.9*x + 2 + np.random.randn(200,1)
plt.plot(x,y,"o")
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
plt.plot(x,y,"o",x_train,lr.predict(x_train),"-r")
```

```python
# POLYNOMIAL REGRESSION
poly = PolynomialFeatures(degree = 2)
print(x_train[0])
x_train = poly.fit_transform(x_train)  # if u set include_bias = False then x**0 value won't be included in the dataset
print(x_train[0])
```

```python
# Train the model on these characteristics
lr.fit(x_train,y_train)
x_plot = np.linspace(-3,3,100).reshape(100,1)
x_plot_poly = poly.fit_transform(x_plot)
y_plot = lr.predict(x_plot_poly)
plt.plot(x,y,"o",x_plot,y_plot,"-r")
```
