```python
import pandas as pd
from pandas_datareader import data
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
x,y = fetch_california_housing(return_X_y=True,as_frame=True)
```

```python
print(x.head())
print(len(x.iloc[:,0]))
```

```python
y.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.1,random_state=3)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)

y_pred = dtr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
dtr.feature_importances_
```

```python
z_train , z_test , y_train , y_test = train_test_split(z,y,test_size=0.1,random_state=5)
```

```python
dtr2 = DecisionTreeRegressor(splitter='random',min_samples_split=25)
dtr2.fit(z_train.values.reshape(-1,1),y_train.values)

y_pred2 = dtr2.predict(z_test.values.reshape(-1,1))
```

```python
r2_score(y_test.values,y_pred2)
```

```python
x_plot = np.linspace(0,9,400).reshape(-1,1)
y_plot = dtr2.predict(x_plot)
```

```python
plt.plot(z.values,y.values,"ob",markersize=0.5)
plt.plot(x_plot,y_plot,color="red")
```
