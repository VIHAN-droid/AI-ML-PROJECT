# RANDOM FOREST CLASSIFICATION

```python
import numpy as np
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
x,y = make_circles(n_samples=500,factor=0.1,noise=0.40,random_state=6)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_test,dt.predict(x_test),clf=dt)
```

```python
from sklearn.metrics import accuracy_score
y_pred = dt.predict(x_test)
accuracy_score(y_test,y_pred)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)

from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_test,rf.predict(x_test),clf=rf)
```

```python
y_pred = rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

# RANDOM FOREST REGRESSION

```python
n_train = 150
n_test = 1000
noise = 0.1

# Generate data
def f(x):
    x = x.ravel()
    return np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples) * 10 - 5
    X = np.sort(X).ravel()
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2)\
        + np.random.normal(0.0, noise, n_samples)
    X = X.reshape((n_samples, 1))

    return X, y

X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)

plt.figure(figsize=(10,6))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.xlim([-5, 5])
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)
d_predict = dt.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, d_predict, "g", lw=2)
plt.xlim([-5, 5])
plt.title("Decision tree, MSE = %.2f"
          % np.sum((y_test - d_predict) ** 2))
```

```python
# USING RF REGRESSOR
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
rf_predict = rf.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "r")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, rf_predict, "g", lw=2)
plt.xlim([-5, 5])
plt.title("Random forest, MSE = %.2f" % np.sum((y_test - rf_predict) ** 2));
```
