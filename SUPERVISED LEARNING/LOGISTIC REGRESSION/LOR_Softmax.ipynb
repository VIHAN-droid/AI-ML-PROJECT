```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
```

```python
# import the encoding module to encode the dataset
from sklearn.preprocessing import LabelEncoder
```

```python
x,y = load_wine(return_X_y=True , as_frame=True)
print(x)
print("\n" , "*" * 100 , "\n")
print(y)
```

```python
x = x[['alcohol' , 'malic_acid']]\
# so there are 2 inputs and output is 0,1,2 i.e there are 3 classes
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.1,random_state=1)

sor = LogisticRegression(multi_class="multinomial")
sor.fit(x_train,y_train)
```

```python
query = pd.DataFrame({"alcohol" : [13.97] , "malic_acid" : [3.54]})
print(query, "\n\n")
print("The predicted output is : ", sor.predict(query))
prob = sor.predict_proba(query)
print("The probabilities are : ",prob)
```

```python
plt.plot([0,1,2] , prob[0] , "or")
plt.xlabel("Class")
plt.ylabel("Probabilities")
plt.ylim(0,0.6)
```
